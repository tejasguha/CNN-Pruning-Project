```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mag = pd.read_csv('results_mag.csv')
dep = pd.read_csv("results_depgraph.csv")
thinet = pd.read_csv("results_thinet.csv")
```

```python
mag = mag.sort_values(by='man_sparsity')
plt.plot(mag['man_sparsity']*100, mag['man_acc'], color='blue', label='Magnitude Based Pruning');

dep = dep.sort_values(by='sparsity')
plt.plot(dep["sparsity"], dep['accuracy'], color='orange', label='DepGraph Pruning');

thinet = thinet.sort_values(by='Sparsity')
plt.plot(thinet["Sparsity"]*100, thinet['Accuracy']*100, color='black', label='Thinet Pruning');


plt.ylim((40, 80));

plt.ylabel('Validation Accuracy');
plt.xlabel('Sparsity');
plt.legend()

plt.title("Pareto Graphs of Attempted Pruning Methods");
```
