```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
reg_3 = pd.read_csv('./data/reg_1e-3_pareto_frontiers.csv')
reg_5 = pd.read_csv('./data/reg_1e-5_pareto_frontiers.csv')
reg_7 = pd.read_csv('./data/reg_1e-7_pareto_frontiers.csv')

reg_3 = reg_3.sort_values(by='sparsity')
reg_5 = reg_5.sort_values(by='sparsity')
reg_7 = reg_7.sort_values(by='sparsity')
```

```python
plt.plot(reg_3['sparsity'], reg_3['accuracy'], color='blue', label='Reg=1e-3');
reg_3 = reg_3.sort_values(by='man_sparsity')
plt.plot(reg_3['man_sparsity'], reg_3['man_acc'], color='blue', linestyle='--');

plt.plot(reg_5['sparsity'], reg_5['accuracy'], color='orange', label='Reg=1e-5');
reg_5 = reg_5.sort_values(by='man_sparsity')
plt.plot(reg_5['man_sparsity'], reg_5['man_acc'], color='orange', linestyle='--');

plt.plot(reg_7['sparsity'], reg_7['accuracy'], color='black', label='Reg=1e-7');
reg_7 = reg_7.sort_values(by='man_sparsity')
plt.plot(reg_7['man_sparsity'], reg_7['man_acc'], color='black', linestyle='--');

plt.ylim((40, 80));

plt.ylabel('Validation Accuracy');
plt.xlabel('Sparsity');
plt.legend()

plt.title("Performance of Magnitude Based Pruning with Different L2 Regularization");
```
